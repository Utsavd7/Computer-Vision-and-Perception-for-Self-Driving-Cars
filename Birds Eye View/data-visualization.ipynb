# Data Visualization
Visualize the Cam2BEV data

# Training Data Visualization

```python
import random
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
TRAIN_DIR = '../input/semantic-segmentation-bev/cam2bev-data-master-1_FRLR/1_FRLR/train/'

BEV_DIR = TRAIN_DIR + 'bev/bev/'
BEV_OCCLUSION_DIR = TRAIN_DIR + 'bev+occlusion/bev+occlusion/'
HOMOGRAPHY_DIR = TRAIN_DIR + 'homography/homography/'

FRONT_DIR = TRAIN_DIR + 'front/front/'
LEFT_DIR = TRAIN_DIR + 'left/left/'
REAR_DIR = TRAIN_DIR + 'rear/rear/'
RIGHT_DIR = TRAIN_DIR + 'right/right/'
```

```python
n_files = 10
files = random.sample(os.listdir(BEV_DIR), n_files)
```

```python
directories = [FRONT_DIR, REAR_DIR, LEFT_DIR, RIGHT_DIR, BEV_DIR, BEV_OCCLUSION_DIR, HOMOGRAPHY_DIR]
titles = ["Front Image", "Rear Image", "Left Image", "Right Image", "BEV Image", "BEV Occlusion Image", "Homography Image"]
```

```python
for i, file in enumerate(files):
    plt.figure(figsize=(25,7))
    plt.tight_layout()
    for i, directory in enumerate(directories):
        ax = plt.subplot(2, 4, i + 1)
        ax.set_title(titles[i])
        img = mpimg.imread(directory + file)
        plt.axis('off')
        plt.imshow(img)

    plt.show()
    print("\n\n\n\n\n\n")
```
